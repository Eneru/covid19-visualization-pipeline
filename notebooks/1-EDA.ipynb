# Import and load data

```python
# notebooks/EDA.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load transformed data
df = pd.read_csv("../data/processed/compact_cleaned.csv", parse_dates=["date"])
```

# Show graphs

## Cases per 100k over time

```python
fig = px.line(df, x="date", y="new_cases_per_100k", color="country",
              title="New COVID-19 Cases per 100k People")
fig.show()
```

## 7-day rolling average of new cases

```python
plt.figure(figsize=(12,6))
for country in df["country"].unique():
    subset = df[df["country"] == country]
    plt.plot(subset["date"], subset["new_cases_7day_avg"], label=country)
plt.legend()
plt.title("7-Day Average of New COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("New Cases (7-day avg)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
